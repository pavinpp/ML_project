```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

class RMSELoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.mse = nn.MSELoss()
    def forward(self, yhat, y):
        return torch.sqrt(self.mse(yhat, y) + 1e-8)  # small epsilon for stability

# =========================
# 1) Clean data & time splits
# =========================

# Keep only non-negative Y_rice
xy_drop_na = xy[xy["Y_rice"] >= 0].copy()

# Remove inf and NaN in all columns
xy_drop_na = xy_drop_na.replace([np.inf, -np.inf], np.nan).dropna()

target_col = "Y_rice"

# remove all Y_* target columns
remove_targets = [c for c in xy.columns if c.startswith("Y_")]

# remove all area columns
remove_areas = [c for c in xy.columns if c.startswith("area_")]

# build feature list
feature_cols = [
    c for c in xy.columns
    if c not in remove_targets
    and c not in remove_areas
    and c != target_col
]

print("Removed target columns:", remove_targets)
print("Removed area columns:", len(remove_areas))
print("Final feature count:", len(feature_cols))

# Time-based split
test_xy  = xy_drop_na[xy_drop_na["year"] >= 2020].copy()
train_val_xy = xy_drop_na[xy_drop_na["year"] < 2020].copy()

train_xy = train_val_xy[train_val_xy["year"] < 2016].copy()
val_xy   = train_val_xy[train_val_xy["year"] >= 2016].copy()

print("Rows -> train:", len(train_xy), "val:", len(val_xy), "test:", len(test_xy))

# =========================
# 2) Features / target & scaling
# =========================

target_col = "Y_rice"
# Only drop target; keep 'year' as a feature
non_feature_cols = ["Y_rice"]

feature_cols = [c for c in xy_drop_na.columns if c not in non_feature_cols]
print("Number of features:", len(feature_cols))

X_train_df = train_xy[feature_cols].copy()
y_train_df = train_xy[[target_col]].copy()

X_val_df   = val_xy[feature_cols].copy()
y_val_df   = val_xy[[target_col]].copy()

X_test_df  = test_xy[feature_cols].copy()
y_test_df  = test_xy[[target_col]].copy()

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_df).astype(np.float32)
X_val   = scaler.transform(X_val_df).astype(np.float32)
X_test  = scaler.transform(X_test_df).astype(np.float32)

y_train = y_train_df.to_numpy().astype(np.float32)
y_val   = y_val_df.to_numpy().astype(np.float32)
y_test  = y_test_df.to_numpy().astype(np.float32)

# =========================
# 3) Sanity check for NaN / inf
# =========================

def check_arr(name, arr):
    print(
        f"{name} -> shape: {arr.shape}, "
        f"NaN: {np.isnan(arr).any()}, "
        f"inf: {np.isinf(arr).any()}"
    )

check_arr("X_train", X_train)
check_arr("y_train", y_train)
check_arr("X_val",   X_val)
check_arr("y_val",   y_val)
check_arr("X_test",  X_test)
check_arr("y_test",  y_test)

# If any of those show True for NaN/inf, that's the source of NaN loss.

# =========================
# 4) PyTorch datasets + loaders
# =========================

train_dataset = TensorDataset(torch.FloatTensor(X_train), torch.FloatTensor(y_train))
val_dataset   = TensorDataset(torch.FloatTensor(X_val),  torch.FloatTensor(y_val))

# test has only features (no labels)
test_dataset  = TensorDataset(torch.FloatTensor(X_test))

batch_size = 64

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# =========================
# 5) Define neural network
# =========================

n_neuron   = 64
N_layers   = 3
activation = "ReLU"
learning_rate = 0.001
num_epochs = 200

layers_list = [nn.Linear(X_train.shape[1], n_neuron),
               getattr(nn, activation)()]

for _ in range(N_layers - 1):
    layers_list.append(nn.Linear(n_neuron, n_neuron))
    layers_list.append(getattr(nn, activation)())

layers_list.append(nn.Linear(n_neuron, 1))  # output layer (Y_rice)

model = nn.Sequential(*layers_list).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.MSELoss()

print(model)

# =========================
# 6) Training loop with early stopping
# =========================

criterion = RMSELoss()     

best_state_dict = model.state_dict()
best_val_loss = float("inf")
patience = 20
patience_counter = 0

train_losses = []
val_losses = []

for epoch in range(num_epochs):
    # ---- Train ----
    model.train()
    running_train_loss = 0.0

    for batch_X, batch_y in train_loader:
        batch_X = batch_X.to(device)
        batch_y = batch_y.to(device)

        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)   # <-- RMSE
        loss.backward()
        optimizer.step()

        running_train_loss += loss.item()

    train_loss = running_train_loss / len(train_loader)

    # ---- Validation ----
    model.eval()
    running_val_loss = 0.0
    with torch.no_grad():
        for batch_X, batch_y in val_loader:
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)   # <-- RMSE
            running_val_loss += loss.item()

    val_loss = running_val_loss / len(val_loader)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    print(f"Epoch {epoch+1}/{num_epochs} "
          f"Train RMSE: {train_loss:.4f}, Val RMSE: {val_loss:.4f}")

    # Early Stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_state_dict = model.state_dict()
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

model.load_state_dict(best_state_dict)


# =========================
# 7) Plot training history
# =========================

def plot_history(train_losses, val_losses):
    plt.figure(figsize=(8,5))
    plt.plot(train_losses, label="Train loss (MSE)")
    plt.plot(val_losses, label="Val loss (MSE)")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training / Validation Loss History")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_history(train_losses, val_losses)

# =========================
# 8) Predict on test set
# =========================
model.eval()
y_pred_test = []

with torch.no_grad():
    for (batch_X,) in test_loader:      # unpack single-tensor batch
        batch_X = batch_X.to(device)
        preds = model(batch_X).cpu().numpy().flatten()
        y_pred_test.extend(preds)

y_pred_test = np.array(y_pred_test, dtype=np.float32)

# ---- Compute metrics ----
from sklearn.metrics import root_mean_squared_error, r2_score

from sklearn.metrics import root_mean_squared_error, r2_score

rmse = root_mean_squared_error(y_test, y_pred_test)
r2   = r2_score(y_test, y_pred_test)

print("RMSE:", rmse)
print("R²:", r2)
```


## Hyperparameter tuning with Optuna

In this section we use Optuna to tune the neural network hyperparameters (number of layers, neurons, dropout, learning rate, and batch size).
We then compare the tuned model performance against the initial neural network using RMSE and R² on the test set.

```python


import optuna

# If Optuna is not installed, uncomment and run the following line once:
# !pip install optuna

def create_model(trial, input_dim):
    """Create a feedforward network whose size/depth are tuned by Optuna."""
    n_layers = trial.suggest_int("n_layers", 1, 3)
    n_neurons = trial.suggest_int("n_neurons", 64, 512, step=64)
    dropout_rate = trial.suggest_float("dropout", 0.0, 0.4)
    
    layers = []
    in_features = input_dim
    for _ in range(n_layers):
        layers.append(nn.Linear(in_features, n_neurons))
        layers.append(nn.ReLU())
        if dropout_rate > 0:
            layers.append(nn.Dropout(dropout_rate))
        in_features = n_neurons
    layers.append(nn.Linear(in_features, 1))
    return nn.Sequential(*layers).to(device)


def objective(trial):
    # Hyperparameters
    lr = trial.suggest_float("lr", 1e-4, 1e-2, log=True)
    batch_size = trial.suggest_categorical("batch_size", [8, 16, 32, 64])
    max_epochs = 80
    patience = 10
    
    # Datasets / loaders
    train_dataset = TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train))
    val_dataset   = TensorDataset(torch.from_numpy(X_val),   torch.from_numpy(y_val))
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False)
    
    model = create_model(trial, X_train.shape[1])
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = RMSELoss()
    
    best_val_loss = float("inf")
    patience_counter = 0
    
    for epoch in range(max_epochs):
        # ---- Train ----
        model.train()
        running_train_loss = 0.0
        for batch_X, batch_y in train_loader:
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)
            
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            
            running_train_loss += loss.item()
        
        # ---- Validation ----
        model.eval()
        val_losses = []
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X = batch_X.to(device)
                batch_y = batch_y.to(device)
                outputs = model(batch_X)
                vloss = criterion(outputs, batch_y)
                val_losses.append(vloss.item())
        
        mean_val_loss = float(np.mean(val_losses))
        
        # Early stopping on validation RMSE
        if mean_val_loss < best_val_loss:
            best_val_loss = mean_val_loss
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                break
    
    # Optuna will minimize this value
    return best_val_loss


# Run Optuna study
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=30, show_progress_bar=False)

print("Number of finished trials:", len(study.trials))
print("Best trial value (val RMSE):", study.best_value)
print("Best hyperparameters:")
for k, v in study.best_trial.params.items():
    print(f"  {k}: {v}")
```

```python

# Train a new model using the best hyperparameters found by Optuna
best_params = study.best_trial.params

tuned_batch_size = best_params.get("batch_size", 64)
tuned_lr = best_params.get("lr", 1e-3)
tuned_n_layers = best_params.get("n_layers", 2)
tuned_n_neurons = best_params.get("n_neurons", 128)
tuned_dropout = best_params.get("dropout", 0.0)

print("Training tuned model with:")
print("  batch_size:", tuned_batch_size)
print("  lr:", tuned_lr)
print("  n_layers:", tuned_n_layers)
print("  n_neurons:", tuned_n_neurons)
print("  dropout:", tuned_dropout)

# Datasets / loaders (train + val are kept separate, early stopping on val)
train_dataset = TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train))
val_dataset   = TensorDataset(torch.from_numpy(X_val),   torch.from_numpy(y_val))
test_dataset  = TensorDataset(torch.from_numpy(X_test))

train_loader = DataLoader(train_dataset, batch_size=tuned_batch_size, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=tuned_batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=tuned_batch_size, shuffle=False)

# Build tuned model
def build_tuned_model(input_dim):
    layers = []
    in_features = input_dim
    for _ in range(tuned_n_layers):
        layers.append(nn.Linear(in_features, tuned_n_neurons))
        layers.append(nn.ReLU())
        if tuned_dropout > 0:
            layers.append(nn.Dropout(tuned_dropout))
        in_features = tuned_n_neurons
    layers.append(nn.Linear(in_features, 1))
    return nn.Sequential(*layers).to(device)

tuned_model = build_tuned_model(X_train.shape[1])
optimizer = torch.optim.Adam(tuned_model.parameters(), lr=tuned_lr)
criterion = RMSELoss()

max_epochs = 200
patience = 20
best_val_loss = float("inf")
patience_counter = 0
best_state_dict_tuned = tuned_model.state_dict()

tuned_train_losses = []
tuned_val_losses = []

for epoch in range(max_epochs):
    # Train
    tuned_model.train()
    running_train_loss = 0.0
    for batch_X, batch_y in train_loader:
        batch_X = batch_X.to(device)
        batch_y = batch_y.to(device)
        
        optimizer.zero_grad()
        outputs = tuned_model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        
        running_train_loss += loss.item()
    
    train_loss = running_train_loss / len(train_loader)
    
    # Validation
    tuned_model.eval()
    val_losses = []
    with torch.no_grad():
        for batch_X, batch_y in val_loader:
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)
            outputs = tuned_model(batch_X)
            vloss = criterion(outputs, batch_y)
            val_losses.append(vloss.item())
    
    val_loss = float(np.mean(val_losses))
    
    tuned_train_losses.append(train_loss)
    tuned_val_losses.append(val_loss)
    
    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_state_dict_tuned = tuned_model.state_dict()
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"[Tuned] Early stopping at epoch {epoch+1}")
            break

# Load best weights
tuned_model.load_state_dict(best_state_dict_tuned)

# Evaluate tuned model on test set
tuned_model.eval()
y_pred_test_tuned = []

with torch.no_grad():
    for (batch_X,) in test_loader:
        batch_X = batch_X.to(device)
        preds = tuned_model(batch_X).cpu().numpy().reshape(-1)
        y_pred_test_tuned.extend(preds)

y_pred_test_tuned = np.array(y_pred_test_tuned, dtype=np.float32)

from sklearn.metrics import root_mean_squared_error, r2_score

rmse_tuned = root_mean_squared_error(y_test, y_pred_test_tuned)
r2_tuned   = r2_score(y_test, y_pred_test_tuned)

print("Tuned model RMSE:", rmse_tuned)
print("Tuned model R²:", r2_tuned)
```
