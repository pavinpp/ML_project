```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_parquet('x_features.parquet')
```

```python
label_yield = pd.read_parquet("label_yield.parquet")
label_yield.head()


label_yield['item'] = label_yield['item'].str.replace(r'[^0-9a-zA-Z ]', '', regex=True)
label_yield['item'] = label_yield['item'].str.replace(" ", "_").str.lower()
label_yield
```

```python
import pandas as pd


label_yield['year'] = pd.to_datetime(label_yield['year'])
label_yield['year'] = label_yield['year'].dt.year

Y = label_yield.pivot_table(
    index=['year','area'],  # row index
    columns='item',         # each item becomes a column
    values='label'          # values
).reset_index()
```

```python
Y.columns = ['year', 'area', 'Y_bananas', 'Y_barley', 'Y_cassava_fresh',
       'Y_cucumbers_and_gherkins', 'Y_maize_corn', 'Y_oil_palm_fruit',
       'Y_other_vegetables_fresh_nec', 'Y_potatoes', 'Y_rice', 'Y_soya_beans',
       'Y_sugar_beet', 'Y_sugar_cane', 'Y_tomatoes', 'Y_watermelons', 'Y_wheat']
```

```python
XY = X.merge(Y, on=['year', 'area'])
```
